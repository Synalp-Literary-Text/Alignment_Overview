```python
import requests
import zipfile
import io
import os

# GitHub repository details
repo_owner = "Synalp-Literary-Text"
repo_name = "Alignment_Overview"
branch = "main"
directory_path = "src/coprus_EN_quote_FR_full"

# Construct the GitHub URL for downloading the ZIP archive
zip_url = f"https://github.com/{repo_owner}/{repo_name}/archive/{branch}.zip"

# Download the ZIP file
print(f"Downloading {zip_url} ...")
response = requests.get(zip_url, stream=True)
if response.status_code == 200:
    with zipfile.ZipFile(io.BytesIO(response.content)) as zip_ref:
        extract_path = f"{repo_name}-{branch}/{directory_path}"
        
        # Extract only the specified directory
        for file in zip_ref.namelist():
            if file.startswith(extract_path):
                zip_ref.extract(file, ".")

        # Move extracted files to a cleaner folder structure
        extracted_folder = f"{repo_name}-{branch}/{directory_path}"
        if os.path.exists(directory_path):
            os.rename(extracted_folder, f"{directory_path}_extracted")

        print(f"Extracted '{directory_path}' successfully!")
else:
    print(f"Failed to download ZIP file. Status code: {response.status_code}")
```
